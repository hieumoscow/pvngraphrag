```python
from azure.core.credentials import AzureKeyCredential
from azure.ai.documentintelligence import DocumentIntelligenceClient
from azure.ai.documentintelligence.models import AnalyzeDocumentRequest, AnalyzeResult
from azure.identity import DefaultAzureCredential
from azure.ai.documentintelligence.models import ContentFormat

def analyse_file_via_doci(file_path, output_path):
    endpoint = "https://graphrag-di.cognitiveservices.azure.com/"
    credential = AzureKeyCredential("9ngq9aMYWwqn9o9T9H0OpdpZZl7ok7nO6QMNOOPEUPUB7AIoOAyfJQQJ99ALAC3pKaRXJ3w3AAALACOGfMVp")
    document_intelligence_client = DocumentIntelligenceClient(endpoint, credential, api_version="2024-11-30")

    # path_to_sample_documents = "/Users/hieunguyennhu/Downloads/prugraphrag/docs/input/C2024_105_Medical_Revision.pdf"
    with open(file_path, "rb") as f:
        file_content = f.read()
        docbytes64 = bytes(file_content)

        poller = document_intelligence_client.begin_analyze_document(
            "prebuilt-layout",
            analyze_request=AnalyzeDocumentRequest(bytes_source=docbytes64),
            output_content_format=ContentFormat.MARKDOWN,
        )
    result: AnalyzeResult = poller.result()

    print(f"Here's the full content in format {result.content_format}:\n")
    print(result.content)

    # store the result in a md file
    # output_path = "/Users/hieunguyennhu/Downloads/prugraphrag/docs/output/C2024_105_Medical_Revision.md"
    with open(output_path, "w", encoding="utf-8") as f:
        f.write(result.content)
    print(f"\nFull content has been written to {output_path}")
```

```python
# iterate through all files in input folder and run the analysis
import os
input_folder = "C:\\Users\hieunhu\Downloads\pvngraphrag\docs\original_files"
output_folder = "C:\\Users\hieunhu\Downloads\pvngraphrag\docs\input"
for file in os.listdir(input_folder):
    if file.endswith(".pdf"):
        file_path = os.path.join(input_folder, file)
        output_path = os.path.join(output_folder, file.replace(".pdf", ".md"))
        print(f"Analyzing {file_path} and saving to {output_path}")
        analyse_file_via_doci(file_path, output_path)
    elif file.endswith(".docx"):
        file_path = os.path.join(input_folder, file)
        output_path = os.path.join(output_folder, file.replace(".docx", ".md"))
        print(f"Analyzing {file_path} and saving to {output_path}")
        analyse_file_via_doci(file_path, output_path)
```
